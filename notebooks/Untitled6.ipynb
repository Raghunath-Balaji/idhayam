```python
PATH = '/dataset/ptb-xl/ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.1/'

df = pd.read_csv(PATH + 'ptbxl_database.csv', index_col='ecg_id')
df.scp_codes = df.scp_codes.apply(ast.literal_eval)

print("Total patients:", len(df))
df.head()
```

```python
scp = pd.read_csv(PATH + 'scp_statements.csv', index_col=0)

# PTB-XL's 5 diagnostic superclasses cover everything
SUPERCLASSES = {
    'NORM': 'Normal',
    'MI':   'Myocardial Infarction',   # heart attack / CAD
    'STTC': 'ST-T Change',             # ischemia, CAD
    'CD':   'Conduction Disturbance',  # arrhythmias, heart block
    'HYP':  'Hypertrophy'              # thickened walls
}

def get_multilabels(scp_dict):
    labels = []
    for sc in SUPERCLASSES.keys():
        has_class = any(
            scp.loc[code, 'diagnostic_class'] == sc
            for code in scp_dict.keys()
            if code in scp.index and
            not pd.isna(scp.loc[code, 'diagnostic_class'])
        )
        labels.append(int(has_class))
    return labels

df['multilabel'] = df.scp_codes.apply(get_multilabels)

# check distribution
import numpy as np
labels_array = np.array(df['multilabel'].tolist())
for i, name in enumerate(SUPERCLASSES.values()):
    print(f"{name}: {labels_array[:, i].sum()} patients")
```

```python
def load_ecg(filename):
    record = wfdb.rdrecord(PATH + filename)
    return record.p_signal.astype(np.float32)

cad_idx = df[df.label == 1].index[0]
normal_idx = df[df.label == 0].index[0]

cad_ecg = load_ecg(df.loc[cad_idx, 'filename_lr'])
normal_ecg = load_ecg(df.loc[normal_idx, 'filename_lr'])

fig, axes = plt.subplots(2, 1, figsize=(15, 6))
axes[0].plot(cad_ecg[:, 1], color='red', linewidth=0.8)
axes[0].set_title('CAD patient — Lead II')
axes[0].set_xlabel('timesteps')
axes[0].set_ylabel('mV')

axes[1].plot(normal_ecg[:, 1], color='green', linewidth=0.8)
axes[1].set_title('Normal patient — Lead II')
axes[1].set_xlabel('timesteps')
axes[1].set_ylabel('mV')

plt.tight_layout()
plt.show()
```

```python
print("Loading ECG signals — takes 2-3 minutes...")

X_ecg = []
valid_idx = []

for ecg_id, row in df.iterrows():
    try:
        signal = load_ecg(row['filename_lr'])
        X_ecg.append(signal)
        valid_idx.append(ecg_id)
    except Exception as e:
        pass

X_ecg = np.array(X_ecg)
y_ecg = np.array(df.loc[valid_idx, 'multilabel'].tolist())  # .tolist() not .values

print("X shape:", X_ecg.shape)   # should be (21837, 1000, 12)
print("y shape:", y_ecg.shape)   # should be (21837, 5)
```

```python
def butterworth_bandpass(signal, lowcut=0.5, highcut=40, fs=100, order=4):
    nyq = fs / 2
    b, a = butter(order, [lowcut/nyq, highcut/nyq], btype='band')
    return filtfilt(b, a, signal, axis=0)

def preprocess_ecg(signal):
    filtered = butterworth_bandpass(signal)
    mean = filtered.mean(axis=0, keepdims=True)
    std = filtered.std(axis=0, keepdims=True) + 1e-8
    return (filtered - mean) / std

print("Preprocessing — takes a few minutes...")
X_ecg = np.array([preprocess_ecg(x) for x in X_ecg])
print("Done:", X_ecg.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_ecg, y_ecg,
    test_size=0.2,
    random_state=42,
)

print("Train:", X_train.shape)
print("Test: ", X_test.shape)
print("CAD in train:", y_train.sum())
print("CAD in test: ", y_test.sum())
```

```python
def residual_block(x, filters, kernel_size=5):
    out = tf.keras.layers.Conv1D(filters, kernel_size, padding='same')(x)
    out = tf.keras.layers.BatchNormalization()(out)
    out = tf.keras.layers.Activation('relu')(out)
    out = tf.keras.layers.Conv1D(filters, kernel_size, padding='same')(out)
    out = tf.keras.layers.BatchNormalization()(out)

    if x.shape[-1] != filters:
        x = tf.keras.layers.Conv1D(filters, 1, padding='same')(x)

    out = tf.keras.layers.Add()([x, out])
    out = tf.keras.layers.Activation('relu')(out)
    return out

def build_ecg_model(input_shape=(1000, 12)):
    inputs = tf.keras.Input(shape=input_shape)

    x = tf.keras.layers.Conv1D(32, 7, padding='same', activation='relu')(inputs)
    x = tf.keras.layers.MaxPooling1D(2)(x)

    x = residual_block(x, 32)
    x = tf.keras.layers.MaxPooling1D(2)(x)

    x = residual_block(x, 64)
    x = tf.keras.layers.MaxPooling1D(2)(x)

    x = residual_block(x, 128)
    x = tf.keras.layers.GlobalAveragePooling1D()(x)

    x = tf.keras.layers.Dense(64, activation='relu', name='embedding')(x)
    x = tf.keras.layers.Dropout(0.3)(x)

    output = tf.keras.layers.Dense(5, activation='sigmoid')(x)

    return tf.keras.Model(inputs, output)

ecg_model = build_ecg_model()
ecg_model.summary()
```

```python
from google.colab import drive
drive.mount('/content/drive')
os.makedirs('/content/drive/MyDrive/heart_project/models', exist_ok=True)

ecg_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor='val_auc', patience=5,
        restore_best_weights=True, mode='max'
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor='val_auc', factor=0.5,
        patience=3, mode='max', verbose=1
    ),
    tf.keras.callbacks.ModelCheckpoint(
        '/content/drive/MyDrive/heart_project/models/ecg_best.keras',
        monitor='val_auc', save_best_only=True, mode='max'
    )
]




history = ecg_model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)
```

```python
y_pred = ecg_model.predict(X_test)  # shape (N, 5)

conditions = ['NORM', 'MI', 'STTC', 'CD', 'HYP']
full_names = {
    'NORM': 'Normal',
    'MI':   'Myocardial Infarction',
    'STTC': 'ST-T Change',
    'CD':   'Conduction Disturbance',
    'HYP':  'Hypertrophy'
}

print("=== Per-condition AUC ===")
for i, cond in enumerate(conditions):
    # skip conditions with no positive cases in test set
    if y_test[:, i].sum() == 0:
        print(f"{full_names[cond]:<30} no positive cases in test set")
        continue
    auc = roc_auc_score(y_test[:, i], y_pred[:, i])
    pos = y_test[:, i].sum()
    print(f"{full_names[cond]:<30} AUC: {auc:.3f}  ({int(pos)} positive cases)")

print("\n=== Per-condition Classification Report ===")
for i, cond in enumerate(conditions):
    if y_test[:, i].sum() == 0:
        continue
    print(f"\n--- {full_names[cond]} ---")
    print(classification_report(
        y_test[:, i],
        (y_pred[:, i] > 0.3).astype(int),
        zero_division=0
    ))
```
